```python
# Converting the units of the age column from days to years
df['age'] = df['age'] / 365.25

# Cleaning dataset to filter out erroneous values of ap_hi and ap_lo
# Calculate BMI
df["bmi"] = df["weight"] / ((df["height"] / 100) * (df["height"] / 100))

# Filter out erroneous values of ap_hi and ap_lo
df = df[(df["ap_hi"] < 200) & (df["ap_hi"] > 50) & (df["ap_lo"] < 200) & (df["ap_lo"] > 50)]
# Filter out erroneous values of BMI
df = df[(df["bmi"] < 40) & (df["bmi"] > 10)]

# Drop the "index" column
df = df.drop("index", axis=1)
# Set the "id" column as the new index
df = df.set_index("id")
```

## Criteria for Selecting Marginal Values as 50 and 200:
These thresholds for ap_low and ap_hi as are chosen based on typical ranges for blood pressure readings in adults and are aimed at filtering out extreme values and impossible measurements while retaining a reasonable range within the dataset.
