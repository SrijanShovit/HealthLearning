## 4. Investigate data type of features and labels and chose any better option for a particular column for data type if possible

```python
df.info() # displays information about the dataset, with respective datatypes for each column
```

Here we identified Columns that Need Encoding and Handling Columns:

-> Category: This column represents categorical data. Encoding is necessary. -> Using One-Hot Encoding.

-> Sex: This column represents binary categorical data. -> Using Binary Encoding via Label Encoding.

-> Columns that Need Handling for Missing Values: ALB, ALP, ALT, CHOL, PROT

```python
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
```

```python
# Print info to confirm the drop
print("DataFrame info after dropping Unnamed: 0:")
df.info()
```

```python
null_values_before = df.isnull().sum() # Checking for null values before handling
print("Null values before handling:")
print(null_values_before)
```

```python
df = pd.get_dummies(df, columns=['Category'], prefix='Category') # Encode 'Category' using One-Hot Encoding and 'Sex' using Label Encoding
df['Sex'] = LabelEncoder().fit_transform(df['Sex'])
```

```python
columns_with_missing_values = ['ALB', 'ALP', 'ALT', 'CHOL', 'PROT'] # Handling missing values: Impute with the mean
for column in columns_with_missing_values:
    df[column].fillna(df[column].mean(), inplace=True)
```

```python
null_values_after = df.isnull().sum() # Checking for null values after handling
print("Null values after handling:")
print(null_values_after)
```

## 5. Calculate the memory usage differences

```python
memory_usage_before = df.memory_usage(deep=True).sum()
print(f"Memory usage before processing: {memory_usage_before / 1024 ** 2:.2f} MB")
df.info()
```

```python
memory_usage_after = df.memory_usage(deep=True).sum()
print(f"Memory usage after processing: {memory_usage_after / 1024 ** 2:.2f} MB")
```
